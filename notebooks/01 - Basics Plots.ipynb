# Basics Plots

- Making Three Types of Simple Plots
- Changing color, line, and make styke
- Labelling our plots


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Configuando o tamanho e o DPI
plt.rcParams['figure.figsize'] = (4,3)
plt.rcParams['figure.dpi'] = 100
```

## Basic Line Plot

```python
# Basic Line Plot
x = np.arange(1, 10, 0.1)
y = np.sin(x)
```

```python
plt.plot(x, np.sin(x))
```

```python
# Plot with x and y axis label
plt.plot(x, y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
```

```python
# Multiple plots
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

```python
plt.plot(x, np.sin(x), label=r"$y = \sin(x)$")
plt.plot(x, np.cos(x), label=r"$y = \cos(x)$")
plt.legend()
plt.xlabel(r"$x$")
plt.ylabel(r"$y$")
plt.title('Basic Plot')
```

```python
# Colors
plt.plot(x, np.sin(x), label=r"$y = \cos(x)$", color="g")
plt.plot(x, np.cos(x), label=r"$y = \cos(x)$", color="r")
plt.legend()
plt.xlabel(r"$x$")
plt.ylabel(r"$y$")
plt.title('Basic Plot Colors')
```

```python
plt.plot(x, np.sin(x), label=r"$y = \cos(x)$", color="g", linestyle='--')
plt.plot(x, np.cos(x), label=r"$y = \cos(x)$", color="r", marker='o', linestyle='none')
plt.legend()
plt.xlabel(r"$x$")
plt.ylabel(r"$y$")
plt.title('Basic Plot Colors')
```

```python
# Short Format Code
#plt.plot(x, np.cos(x), label=r"$y = \cos(x)$", color="r", marker='o', linestyle='none')
plt.plot(x, np.cos(x), "r+--", label=r"$y = \cos(x)$")
plt.legend()
plt.xlabel(r"$x$")
plt.ylabel(r"$y$")
plt.title('Basic Plot Short Format Code')
```

```python
# Axis Limit. Wiht this we can see limits that we defines to analyze
# Short Format Code
plt.plot(x, np.sin(x), "r--", label=r"$y = \cos(x)$")
plt.plot(x, np.cos(x), "bo", label=r"$y = \cos(x)$")
#if you are interested in the positive values
plt.ylim(0,1)
plt.xlim(0,1)

#plt.xlim(0,1)
plt.legend()
plt.xlabel(r"$x$")
plt.ylabel(r"$y$")
plt.title('Basic Plot Short Format Code')
```

## Histogram

```python
plt.hist(np.random.normal(size=1000))
plt.xlabel(r"$x$")
plt.ylabel(r"$Pr(x)$")
```

# Plot Many Dimensions

```python
#generate a smoothed, gaussian random field
from scipy.ndimage.filters import gaussian_filter
rands2d = gaussian_filter(np.random.normal(size=(512,512)), sigma=10)
```

```python
rands2d.shape
```

```python
plt.imshow(rands2d)
plt.colorbar()
```

```python
plt.imshow(rands2d, cmap='viridis')
plt.colorbar()
```

```python
plt.imshow(rands2d, cmap='magma')
plt.colorbar()
```

```python
plt.imshow(rands2d, cmap='inferno')
plt.colorbar()
```
